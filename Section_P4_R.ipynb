```r
F<-c(20,40,60,80,100)
d<-c(1.4,2.5,4.0,5.3,6.6)
```

```r
plot(F,d,col="blue", main="Hooke's Law: F=kd",xlab="Force (N)", ylab="distance (cm)")
```

```r
relation<-lm(d~I(F))
print(relation)
summary(relation)
```

```r
F_seq <- seq(from=20,to=100, by=0.1)
new_relation<-list(0)
for (i in 1:length(F_seq)) {
    F_new <- data.frame(F=F_seq[i])
    new_relation[i]<-predict(relation, F_new)
}
plot(F,d,col="blue", main="Linear Regression Example",lines(F_seq, new_relation))
```

```r
RSE_Percent = 0.09661/mean(d) * 100
```

```r
RSE_Percent
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```

```r
predict(relation, data.frame(F=55))
```

```r
x<-c(4,6,8,10,12)
y<-c(2370,5460,10310,16250,23860)
```

```r
plot(x,y,col="blue", main="Beam Strength",xlab="Beam Height (in)", ylab="Strength (lbs)")
```

```r
relation2<-lm(y~I(x)+I(x^2))
print(relation2)
summary(relation2)
```

```r
x_seq <- seq(from=4,to=12, by=0.01)
new_relation2<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation2[i]<-predict(relation2, x_new)
}
plot(x,y,col="blue", main="Beam Strength",lines(x_seq, new_relation2),xlab="Beam Height (in)", ylab="Strength (lbs)")
```

```r
RSE_Percent = 107.5/mean(y) * 100
```

```r
RSE_Percent
```

```r
confint(relation2, level=0.95)
```

```r
plot(fitted(relation2), residuals(relation2))
```

```r
predict(relation2, data.frame(x=2))
```

```r
x<-c(1,2,3,4,5,6)
y<-c(40,85,140,200,225,245)
```

```r
plot(x,y,col="blue", main="Engine Performance",xlab="Engine Speed (x1000 rpm)", ylab="Horsepower")
```

```r
relation3<-lm(y~I(x)+I(x^2)+I(x^3))
print(relation3)
summary(relation3)
```

```r
x_seq <- seq(from=1,to=6, by=0.01)
new_relation3<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation3[i]<-predict(relation3, x_new)
}
plot(x,y,col="blue", main="Engine Performance",lines(x_seq, new_relation3),xlab="Engine Speed (x1000 rpm)", ylab="Horsepower")
```

```r
RSE_Percent = 6.614/mean(y) * 100
```

```r
RSE_Percent
```

```r
confint(relation3, level=0.95)
```

```r
plot(fitted(relation3), residuals(relation3))
```

```r
predict(relation3, data.frame(x=4.5))
```
