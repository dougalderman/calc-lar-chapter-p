```r
x<-c(3,6,9,12,15,18,21,24,27,30)
y<-c(61,56,53,55,48,35,36,33,44,23)
```

```r
plot(x,y,col="blue", main="Machine Part Test",xlab="Bending length (in cm, 10x/minute)", ylab="Time to failure (in hours)")
```

```r
relation<-lm(y~I(x))
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=3,to=30, by=0.01)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Machine Part Test",lines(x_seq, new_relation),xlab="Bending length (in cm, 10x/minute)",ylab="Time to failure (in hours)")
```

```r
RSE_Percent = 5.916/mean(y) * 100
print(RSE_Percent)
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```
